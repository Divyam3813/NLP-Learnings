```python
corpus=""" Hello Welcome to my ipynb file's.
Please study regularly! to become expert."""
```

```python
print(corpus)
```

```python
## Convert paragraph into sentences.
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
documents=sent_tokenize(corpus)
```

```python
for sentence in documents:
    print(sentence)
```

```python
type(documents)
```

```python
## Convert paragraph and sentences into words.
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentences in documents:
    print(word_tokenize(sentences))
```

```python
from nltk.tokenize import wordpunct_tokenize ## Treats the punctiation like ' in 's as another word.
```

```python
wordpunct_tokenize(corpus)
```
